```python
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function

import numpy as np
from scipy.special import digamma
from scipy.misc import logsumexp
from scipy.special import gammaln
import matplotlib.pyplot as plt

import os, sys, h5py, time
sys.path.append('../../..')
import mutagenesisfunctions as mf
import bpdev as bd
import helper
```

# Open data as onehot

```python
fam = 'RF00167'
filename = '../../data_marks/%s/%s_red.hdf5'%(fam, fam)
with h5py.File(filename, 'r') as dataset:
    X_data = np.array(dataset['X_data'])
numdata = X_data.shape[0]
X_pos = X_data[:numdata//2, :, 0, :] #include gaps so the profile will learn an alignment
numdata, seqlen, dims = X_pos.shape

x_nla = X_pos
x_nla.shape
```

# Train variational bayes profile mixture model 

It's good practice to run a few times and see what the highest elbo is. Proceed with the parameters that yield highest elbo.  

```python
num_trials = 50         # number of trials to run
K = 50                  # number of states
alpha_strength = 1e-3   # alpha < 1 (sparse), alpha > 1 (dense), alpha=1 (standard)
beta_strength = 1       # beta < 1 (sparse), beta > 1 (dense), beta=1 (standard)
max_iter = 100          # maximum number of iterations
convergence = 1e-8      # convergence criterion
verbose = True

N,L,A = x_nla.shape


alpha_trial = []
beta_trial = []
r_nk_trial = []
elbo_trial = []

for trial in range(num_trials):
    print('Trial %d out of %d'%(trial+1, num_trials))

    # set priors 
    alpha_0 = np.ones(K)*alpha_strength       # 
    beta_0 = np.ones((K,L,A))*beta_strength

    # initial posterior for latent states
    r_nk = np.random.rand(N, K)
    r_nk /= np.sum(r_nk, axis=1, keepdims=True)

    # train model
    old_log_elbo = -1e12
    converged = False
    log_elbo_epochs = []
    for i in range(max_iter):

        # maximization
        alpha_k = alpha_0 + np.sum(r_nk, axis=0)
        beta_kla = beta_0 + np.einsum('nk,nla->kla', r_nk, x_nla)

        # useful expectations
        E_ln_pi_k = digamma(alpha_k) - digamma(np.sum(alpha_k)) # K
        E_ln_phi_kla = digamma(beta_kla) - digamma(np.sum(beta_kla, axis=2, keepdims=True)) # K, L, A

        # expectation
        ln_rho_nk = E_ln_pi_k + np.einsum('nla,kla->nk', x_nla, E_ln_phi_kla)
        ln_sum_rho_nk = logsumexp(ln_rho_nk, axis=1, keepdims=True)
        r_nk = np.exp(ln_rho_nk - ln_sum_rho_nk)

        # Variational lower bound
        E_ln_p_x = np.sum(np.einsum('nk,nla->kla', r_nk, x_nla)*E_ln_phi_kla)
        E_ln_p_z = np.sum(r_nk * E_ln_pi_k)

        ln_C_alpha_0 = gammaln(np.sum(alpha_0)) - np.sum(gammaln(alpha_0))
        E_ln_p_pi = ln_C_alpha_0 + np.sum((alpha_0-1)*E_ln_pi_k)

        ln_C_beta_0 = gammaln(np.sum(beta_0, axis=2, keepdims=True)) - np.sum(gammaln(beta_0), axis=2, keepdims=True)
        E_ln_p_phi = np.sum(ln_C_beta_0 + np.sum((beta_0-1)*E_ln_phi_kla, axis=2, keepdims=True))

        E_ln_q_z = np.sum(r_nk * np.log(r_nk + 1e-7))

        ln_C_alpha_k = gammaln(np.sum(alpha_k)) - np.sum(gammaln(alpha_k))
        E_ln_q_pi = ln_C_alpha_k + np.sum((alpha_k - 1)*E_ln_pi_k)

        ln_C_beta_kla = gammaln(np.sum(beta_kla, axis=2, keepdims=True)) - np.sum(gammaln(beta_kla), axis=2, keepdims=True)
        E_ln_q_phi = np.sum(ln_C_beta_kla + np.sum((beta_kla-1)*E_ln_phi_kla, axis=2, keepdims=True))

        log_elbo = E_ln_p_x + E_ln_p_z + E_ln_p_pi + E_ln_p_phi - E_ln_q_z - E_ln_q_pi - E_ln_q_phi

        # store lower bound
        log_elbo_epochs.append(log_elbo)

        # print progress
        if verbose:
            print("  iteration: %d  log_elbo: %f"%(i, log_elbo))

        # check for convergence
        if (log_elbo - old_log_elbo)/np.abs(log_elbo) < convergence:
            print('Converged after %d iterations'%(i))
            converged = True
            break
        else:
            old_log_elbo = log_elbo

    if not converged:
        print('Did not converge in %d iterations'%(max_iter))

    alpha_trial.append(alpha_k)
    beta_trial.append(beta_kla)
    elbo_trial.append(log_elbo)
    r_nk_trial.append(r_nk)
    
optimal_trial = np.argmax(elbo_trial)
alpha_k = alpha_trial[optimal_trial]
beta_kla = beta_trial[optimal_trial]
r_nk = r_nk_trial[optimal_trial]
print('Optimal log-elbo: %f'%(elbo_trial[optimal_trial]))
```

```python
# plot the ELBO at each epoch --> should monotonically increase if no bugs
plt.figure()
plt.plot(log_elbo_epochs)
plt.xlabel('epochs')
plt.ylabel('ELBO')
plt.show()
```

```python
def dirichlet_mean_std(alpha, axis=0):
    """Calculate the mean and std dev of the Dirichlet distribution with the
       given concentraion parameters."""
    mean = alpha / np.sum(alpha, axis=axis, keepdims=True)
    alpha0 = np.sum(alpha, axis=axis, keepdims=True)
    std = np.sqrt(alpha*(alpha0 - alpha)/(alpha0**2*(alpha0 + 1)))
    return mean, std


# get mean and std of beta distribution with inferred parameters of the latent states (z)
alpha_mean, alpha_std = dirichlet_mean_std(alpha_k, axis=0)

# get mean and std of beta distribution for the profiles
beta_mean, beta_std = dirichlet_mean_std(beta_kla, axis=2)

# get only profiles for states with significant probability mass for latent state
state_index = np.where(alpha_mean > 0.01)[0]
beta_mean = beta_mean[state_index]
beta_std = beta_std[state_index]

#plot population fraction of each state
fig = plt.figure()
plt.plot(alpha_mean)
plt.xlabel("States")
plt.ylabel("Probability")
plt.show()
```

# Sample sequences from the inferred profiles

There are 19 significant states and therefore profiles that the model has predicted. We just need to use the significant betas as profiles to emit sequences. We should sample a profile proportional to its alpha and then emit a sequence

```python
from collections import Counter
np.random.seed(42)

N = numdata #number of sequences to sample

sample_prop = np.copy(alpha_mean[state_index])
sample_prop = sample_prop/np.sum(sample_prop)
sample_k = np.random.choice(len(sample_prop), size=N, p=sample_prop)

cnt_k = Counter()
for sk in sample_k:
    cnt_k[sk] += 1
```

```python
Xnegprofile = []
tottime = time.time()
for sk in cnt_k:
    starttime = time.time()
    Xnegprofile.append(mf.seq_generator_profile(beta_mean[sk], cnt_k[sk], seqlen, dims-1))

    print ('Sequences emitted from profile %s: '%(sk)+ mf.sectotime(time.time() - starttime))

Xnegprofile = np.vstack(Xnegprofile)
print ('Profile sequences emitted:', mf.sectotime(time.time() - tottime))
```
