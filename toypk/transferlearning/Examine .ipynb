```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
filename = 'test_metrics.csv'
df = pd.read_csv(filename, header=0)
df
```

```python
df_perf = df[['modelarch', 'transfer', 'numpos', 'accuracy', 'bp_ppv', 'totscore']]
df_perf
```

```python
t = sb.factorplot(x='numpos', y='accuracy', col='modelarch',
              hue='transfer', data=df_perf, kind='point')
t.set_xticklabels(rotation=30)
t.set_titles("{col_name}")
```

```python
t = sb.factorplot(x='numpos', y='bp_ppv', col='modelarch',
              hue='transfer', data=df_perf, kind='point')
t.set_xticklabels(rotation=30)
t.set_titles("{col_name}")
```

```python
t = sb.factorplot(x='numpos', y='totscore', col='modelarch',
              hue='transfer', data=df_perf, kind='point')
t.set_xticklabels(rotation=30)
t.set_titles("{col_name}")
```
