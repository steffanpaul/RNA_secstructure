```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
filename = 'test_metrics.csv'
df = pd.read_csv(filename, header=0)
```

```python
trnaruns = df.loc[df['data'] == 'trna'][['modelarch', 'data', 'numpos', 'accuracy', 'bp_ppv']]
```

```python
df_perf = df[['modelarch', 'data', 'numpos', 'accuracy', 'bp_ppv', 'totscore']]
```

```python
t = sb.factorplot(x='numpos', y='accuracy', col='data',
              hue='modelarch', data=df_perf, kind='point',
                 yticklabels=[])
t.set_xticklabels(rotation=30)
t.set_titles("{col_name}")
plt.yticks(t.get_yticks(), t.get_yticks() * 100)
```

```python
sb.lineplot(x='numpos', y='accuracy', hue = 'modelarch', data=df_perf)
```

```python
t = sb.factorplot(x='numpos', y='bp_ppv', col='data',
              hue='modelarch', data=df_perf, kind='point')

t.set_xticklabels(rotation=30)
t.set_titles("{col_name}")
```

```python
t = sb.factorplot(x='numpos', y='totscore', col='data',
              hue='modelarch', data=df_perf, kind='point')
t.set_xticklabels(rotation=30)
t.set_titles("{col_name}")
```
