```python
import pandas as pd
import numpy as np

df = pd.read_csv('test_metrics.csv')
df_view = df[['Hidden Layer Size', 'Family', 'NumTrained', 'Accuracy']].sort_values(['Family', 'Hidden Layer Size']).reset_index(drop=True)
```

```python
#retrieve the sequence statistics from the .txt file
stat_vars = ['Family Name', 'Alignment depth', 'Alignment length', 'Mean seq length',
        'Min seq length', 'Max seq length', 'Std seq length']
stat_file = 'zwein_data_details.txt'

fam_stats = {st: [] for st in stat_vars}

with open(stat_file, 'r') as fd:
    found = False
    for line in fd:
        line = line.strip('\n')
        for st in stat_vars:
            if st in line:
                line = line.split()
                fam_stats[st].append(line[-1])

df_stats = pd.DataFrame(fam_stats) 
df_stats = df_stats.apply(pd.to_numeric, errors='ignore')
df_stats = df_stats.drop_duplicates(subset="Family Name")
```

```python
#size_ratio -> depth/mean_length. A deeper alignment and a shorter sequence is preferrable
l = []
for ii in range(len(df_view)):
    f = df_view.iloc[ii, 1]
    l.append(df_stats[df_stats['Family Name'] == f]['Mean seq length'].iloc[0])
L = pd.Series(l)
df_view['Mean seq length'] = L
df_view['Size ratio'] = df_view['NumTrained']/df_view['Mean seq length']
```

```python
df_stats
meltdf= pd.melt(df_stats, id_vars=['Alignment depth', 'Family Name'], var_name="Length type", value_name="Length stat")
```

```python
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure()
sb.set(style='ticks')
sb.catplot(x = 'Size ratio', y='Accuracy', data=df_view, hue='Family',
           col_wrap=3, col='Hidden Layer Size', sharex=True)
```

```python
sb.catplot(x = 'Alignment depth', y='Length stat', data=meltdf, hue='Family Name',
           col='Length type', col_wrap=3, sharex=False)
```
